```python
songs = soup.find_all(class_="title")
songs
```

```python
song_names=[]
```

```python
for a in soup.find_all('cite'):
    song_names.append(a.get_text())
```

```python
song_names
```

```python
artists = soup.find_all(class_="artist")
artists
```

```python
artist= []
```

```python
for a in soup.find_all('em'):
    artist.append(a.get_text())
```

```python
artist
```

```python
df = pd.DataFrame(
    {"songs": song_names,
     "artist": artist})
```
