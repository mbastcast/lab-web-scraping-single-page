```python
from bs4 import BeautifulSoup   ### pip install bs4
import pandas as pd
import requests

```

```python
html = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
r = requests.get('http://www.popvortex.com/music/charts/top-100-songs.php')
r.status_code
```

```python
html = r.content
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup.head()
```

```python
##this is one song
```

```python
songs = soup.find_all(class_="title")
songs
```

```python
song_names=[]
```

```python
for a in soup.find_all('cite'):
    song_names.append(a.get_text())
```

```python
song_names
```

```python
song_names.remove('Billboard Hot 100')
```

```python
song_names.remove('The Voice')
```

```python
song_names
```

```python
artists = soup.find_all(class_="artist")
artists
```

```python
artist= []
```

```python
for a in soup.find_all('em'):
    artist.append(a.get_text())
    
```

```python
artist
```

```python
artist.remove('New Release')
```

```python
len(artist)
```

```python
len(song_names)
```

```python
df = pd.DataFrame(
    {"songs": song_names,
     "artist": artist})
```
